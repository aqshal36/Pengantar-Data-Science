```python
import pandas as pd
df = pd.read_excel('/content/drive/MyDrive/Phyton/tugas 2/Data Tugas Tuton STDA4101-2024.2.xlsx')
```

```python
df = df[['Y', 'LTG']]
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt

```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['LTG'], df['Y'], color='blue', alpha=0.7)
plt.title('Scatter Plot: LTG vs Y')
plt.xlabel('LTG (Triglycerides)')
plt.ylabel('Y (Perkembangan Penyakit Setelah Satu Tahun)')
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
import numpy as np
```

```python
X = df[['LTG']]
y = df['Y']
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
slope = model.coef_[0]
intercept = model.intercept_
r2 = model.score(X, y)
```

```python
print(f"Persamaan regresi: Y = {intercept:.2f} + {slope:.2f} * LTG")
print(f"Koefisien Determinasi (R²): {r2:.4f}")
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Data Asli')
plt.plot(X, model.predict(X), color='red', label='Regresi Linear')
plt.title('Regresi Linear: LTG vs Y')
plt.xlabel('LTG')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```
