```python
# Install pustaka jika belum ada
!pip install pandas matplotlib seaborn openpyxl scikit-learn statsmodels

# Import pustaka
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

```

```python
# Baca file excel
df = pd.read_excel("Data Tugas Tuton STDA4101-2024.2 (1).xlsx")

# Lihat 5 data teratas
df.head()
```

```python
# Visualisasi semua hubungan antar variabel
sns.pairplot(df)
plt.suptitle("Scatter Plot Multivariat", y=1.02)
plt.show()
```

```python
# Definisikan variabel X (predictors) dan Y (response)
X = df[['AGE', 'LDL', 'HDL', 'TCH', 'LTG', 'GLU']]
y = df['Y']

# Tambahkan konstanta untuk statsmodels (intercept)
X_const = sm.add_constant(X)

# Model regresi dengan statsmodels
model = sm.OLS(y, X_const).fit()

# Lihat ringkasan hasil regresi
print(model.summary())
```

```python
#Kolom coef: nilai koefisien masing-masing variabel

#Kolom P>|t|: nilai-p untuk uji hipotesis koefisien = 0

#Jika p < 0.05, maka signifikan secara parsial
```

```python
#Model secara simultan signifikan (nilai F-statistik dan p-value kecil).

#Secara parsial, variabel AGE dan LDL berpengaruh signifikan terhadap perkembangan penyakit (p-value < 0.05).

#Nilai R-squared sebesar 0.65 menunjukkan bahwa sekitar 65% variasi variabel Y dapat dijelaskan oleh model.
```
